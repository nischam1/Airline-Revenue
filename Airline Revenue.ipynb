# 5. Linear Regression

```python
import statsmodels.formula.api as smf


model = smf.ols('TOTAL_Pass ~ TOTAL_Op_Rev', data=Yearly)
model = model.fit()
```

```python
# Predict values
Pass_Pred = model.predict()

# Plot regression against actual data
plt.figure(figsize=(30, 6))
plt.plot(Yearly['TOTAL_Op_Rev'], Yearly['TOTAL_Pass'], 'o')          
plt.plot(Yearly['TOTAL_Op_Rev'], Pass_Pred, 'r', linewidth=2)
plt.xlabel('TOTAL_Op_Rev')
plt.ylabel('TOTAL_Pass')
plt.title('TOTAL_Pass vs TOTAL_Op_Rev')

plt.show()
```

```python
model2 = smf.ols('TOTAL_Flight ~ TOTAL_Op_Rev', data=Yearly)
model2 = model2.fit()
```

```python
# Predict values
Pass_Pred = model2.predict()

# Plot regression against actual data
plt.figure(figsize=(30, 6))
plt.plot(Yearly['TOTAL_Op_Rev'], Yearly['TOTAL_Flight'], 'o')          
plt.plot(Yearly['TOTAL_Op_Rev'], Pass_Pred, 'r', linewidth=2)
plt.xlabel('TOTAL_Op_Rev')
plt.ylabel('TOTAL_Flight')
plt.title('TOTAL_Flight vs TOTAL_Op_Rev')

plt.show()
```
